```python
raw_data = {'year': row_val[2][2:],
           'murder':row_val[4][2:],
           'robber':row_val[5][2:],
           'sex':row_val[6][2:],
            'violence':row_val[7][2:],
            'theft':row_val[8][2:]}

data = pd.DataFrame(raw_data)
data['total'] = total2
```

```python
np_data = np.array(data)
```

```python
data_2016 = np_data[-1]
```

```python
data_2016
```

```python
total_2016 = float(data_2016[-1])
major_2016 = 0
for i in range(len(data_2016)-2):
    major_2016 += float(data_2016[i])
```

```python
major_2016
```

```python
labels = ['기타 형법범죄', '주요 형범범죄\n(살인, 강도, 성폭력, 폭행, 절도)']
ratio = [total_2016-major_2016, major_2016]
explode=(0.0, 0.1)
colors = ['gold', 'lightskyblue']
print(ratio)
```

```python
plt.pie(ratio, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%\n', shadow=True, startangle=90)
plt.title('2016년 형법범죄 비율')
plt.show()
```
